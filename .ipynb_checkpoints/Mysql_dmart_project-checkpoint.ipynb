```python
orders_df = pd.read_sql_query("SELECT * FROM orders", connection)
order_details_df = pd.read_sql_query("SELECT * FROM OrderDetails", connection)
products_df = pd.read_sql_query("SELECT * FROM Products", connection)
```

```python
# 2. Pie Chart: Distribution of orders across different product categories
orders_by_category = pd.merge(orders_df, order_details_df, on='OrderID')
orders_by_category = pd.merge(orders_by_category, products_df, on='ProductID')
orders_by_category = orders_by_category.groupby('Category')['OrderID'].count()
plt.figure(figsize=(8, 8))
plt.pie(orders_by_category, labels=orders_by_category.index, autopct='%1.1f%%')
plt.title('Distribution of Orders by Product Category')
plt.show()
```

```python
# 3. Histogram: Distribution of order quantities
plt.figure(figsize=(10, 6))
plt.hist(order_details_df['Price'], bins=10, edgecolor='black')
plt.title('Distribution of Order price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python

# 1. Bar Plot: Total amount of orders per customer
orders_per_customer = orders_df.groupby('CustomerID')['TotalAmount'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=orders_per_customer, x='CustomerID', y='TotalAmount')
plt.title('Total Amount of Orders per Customer')
plt.xlabel('CustomerID')
plt.ylabel('Total Amount')
plt.show()

```

```python
# 5. Scatter Plot: Relationship between order quantity and total amount
order_details_merged = pd.merge(order_details_df, orders_df, on='OrderID')
plt.figure(figsize=(10, 6))
plt.scatter(order_details_merged['Quantity'], order_details_merged['TotalAmount'])
plt.title('Relationship between Order Quantity and Total Amount')
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Create a boxplot of order amounts
plt.figure(figsize=(10, 6))
sns.boxplot(data=orders_df, y='TotalAmount')
plt.title('Distribution of Order Amounts')
plt.ylabel('Total Amount')
plt.show()
# This boxplot will display the distribution of order amounts, showing the median, quartiles, and any potential outliers. 
# It's a useful visualization for understanding the spread of order amounts and identifying any extreme values.
```
